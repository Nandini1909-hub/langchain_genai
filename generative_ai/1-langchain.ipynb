```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model="groq:llama-3.3-70b-versatile"
)

response = model.invoke("Write me a joke on machine learning algorithms")

print(response.content)
```

##Streaming and Batch

```python
model.invoke("Write me a 500 words paragraph on AI")
```

```python
for chunk in model.stream("Write me a 500 words paragraph on AI"):
    print(chunk.text,end="",flush = True)
```

```python
model.invoke("Why do parrots have colorful feathers?")
```

```python
for chunk in model.stream("Why do parrots have colorful feathers?"):
    print(chunk.text,end="",flush=True)
```

```python
responses = model.batch(
    [
        "Why do parrots have colorful feathers?",
        "What is the capital of France?",
        "What is the square root of 16?"
    ]
)
for r in responses:
    print(r.content)
```

```python
responses = model.batch(
    [
        "Why do parrots have colorful feathers?",
        "What is the capital of France?",
        "What is the square root of 16?"
    ],
    config={
        'max_concurrency':5,
    }
)
```
